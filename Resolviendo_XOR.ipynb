```python
import numpy as np

def escalon(x):
    return 1 if x >= 0 else 0

# Perceptrón OR
def OR(x):
    w = np.array([1,1])
    b = -0.5
    return escalon(np.dot(x, w) + b)

# Perceptrón NAND
def NAND(x):
    w = np.array([-1,-1])
    b = 1.5
    return escalon(np.dot(x, w) + b)

# Perceptrón AND
def AND(x):
    w = np.array([1,1])
    b = -1.5
    return escalon(np.dot(x, w) + b)

# XOR usando los anteriores
def XOR(x):
    salida_or = OR(x)
    salida_nand = NAND(x)
    return AND(np.array([salida_or, salida_nand]))

# Prueba
inputs = np.array([[0,0],[0,1],[1,0],[1,1]])

print("x1 x2 -> XOR")
for x in inputs:
    print(x[0], x[1], "->", XOR(x))
    
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron

# Intentar resolver XOR con un solo Perceptrón usando scikit-learn (similar a SKlearnDiplo)
inputs = np.array([[0,0],[0,1],[1,0],[1,1]])
targets = np.array([0,1,1,0])  # XOR targets

# Modelo
modelo = Perceptron(eta0=0.1, max_iter=1, warm_start=True, random_state=42)
# eta = índice de aprendizaje
# max_iter = no repita entradas tipo 0,0 o 0,1   # Hiperparámetros
# warm_start = recordar valores

epochs = 15  # épocas
errores = []

modelo.partial_fit(inputs, targets, classes=np.array([0,1]))  # partial = mostrar cambios con cada epoca

print("\nPesos Iniciales")
print("Weights:", modelo.coef_)
print("Bias:", modelo.intercept_)

# aprendizaje
for epoch in range(epochs):
    modelo.partial_fit(inputs, targets)

    predicciones = modelo.predict(inputs)
    error = np.sum(predicciones != targets)
    errores.append(error)

    # ver epocas
    print(f"\nÉpoca {epoch+1}")
    print("Predicciones:", predicciones)
    print("Error:", error)
    print("Weights:", modelo.coef_)
    print("Bias:", modelo.intercept_)

    # Verificar condición de paro
    if error == 0:
        print("Condición de paro: Disparada.")
        break

# Finales
print("\nPesos Finales")
print("Weights:", modelo.coef_)
print("Bias:", modelo.intercept_)

plt.figure()
plt.plot(range(1, len(errores)+1), errores, marker='o')
plt.xlabel("Épocas")
plt.ylabel("Número de errores")
plt.title("Error del Perceptrón intentando resolver XOR")
plt.grid()
plt.show()
```

```python
from sklearn.linear_model import Perceptron

modelo = Perceptron(eta0=0.1, max_iter=1, warm_start=True)

# Implementar XOR con perceptrones multicapa usando scikit-learn, entrenando como en la línea
# OR, NAND en capa oculta, AND en salida

# Crear perceptrones con parámetros como en la línea
or_perceptron = Perceptron(eta0=0.001, max_iter=1000, warm_start=True)
nand_perceptron = Perceptron(eta0=0.001, max_iter=1000, warm_start=True)
and_perceptron = Perceptron(eta0=0.001, max_iter=1000, warm_start=True)

# Targets para cada perceptron
or_targets = np.array([0, 1, 1, 1])  # OR
nand_targets = np.array([1, 1, 0, 0])  # NAND
xor_targets = np.array([0, 1, 1, 0])  # XOR

print("\nEntrenando red multicapa para XOR:")

# Entrenar OR y NAND completamente primero (capa oculta fija)
or_perceptron.fit(inputs, or_targets)
nand_perceptron.fit(inputs, nand_targets)

# Obtener salidas de capa oculta (ahora fijas)
or_out = or_perceptron.predict(inputs)
nand_out = nand_perceptron.predict(inputs)
hidden_inputs = np.column_stack((or_out, nand_out))

print("Salidas ocultas fijas:")
print("OR:", or_out)
print("NAND:", nand_out)
print("Hidden inputs:", hidden_inputs)

# Entrenar AND en hidden fijo hasta convergencia
epochs = 300
errores = []

print(or_perceptron.predict(inputs))
print(nand_perceptron.predict(inputs))

for epoch in range(epochs):
    indices = np.random.permutation(len(hidden_inputs))
    X = hidden_inputs[indices]
    y = xor_targets[indices]

    and_perceptron.partial_fit(X, y, classes=[0, 1])

    xor_out = and_perceptron.predict(hidden_inputs)
    error = np.sum(xor_out != xor_targets)
    errores.append(error)

    print(f"Época {epoch+1}: Error = {error}")

    if error == 0:
        print("¡Convergencia lograda!")
        break

print("\nPesos finales OR:")
print("Weights:", or_perceptron.coef_)
print("Bias:", or_perceptron.intercept_)

print("\nPesos finales NAND:")
print("Weights:", nand_perceptron.coef_)
print("Bias:", nand_perceptron.intercept_)

print("\nPesos finales AND:")
print("Weights:", and_perceptron.coef_)
print("Bias:", and_perceptron.intercept_)

print("\nSalidas finales:")
print("OR:", or_out)
print("NAND:", nand_out)
print("XOR:", xor_out)

print("\nComparación con XOR manual:")
print("x1 x2 -> Manual -> Multicapa entrenada")
for i, x in enumerate(inputs):
    manual = XOR(x)
    pred = xor_out[i]
    print(f"{x[0]}  {x[1]}     {manual}            {pred}")

# Gráfica del error por épocas
plt.figure()
plt.plot(range(1, len(errores)+1), errores, marker='o')
plt.xlabel("Épocas")
plt.ylabel("Número de errores")
plt.title("Error de la red multicapa entrenada")
plt.grid()
plt.show()
```
