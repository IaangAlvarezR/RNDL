```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definiendo funcion de activacion
def escalon_unitario(x):
  if x>=0:
    return 1
  else:
    return 0
```

**Compuerta OR**

---



```python
# Cada fila de esta matriz representa un 'registro de entrada' para el perceptrón
inputs = np.array([[0, 0], [0,1], [1,0], [1,1]])

# Los 'targets' son los valores que queremos que el perceptrón de a la salida
# con cada registro de entrada
targets = np.array([0,1,1,1])      #Esta combinación pertenece a la compuerta OR

# En esta variable de tipo lista almacenaremos las salidas del perceptrón con cada registro de entrada
outputs = []

# Modifica el valor de los pesos (weights) y del bias del perceptrón
# para que al hacer la suma ponderada (o combinación lineal) y llamar a la
# función de activación la salida del perceptrón sea igual al target
# para cada registro de entrada
weights = np.array([-1,1])    # El vector de pesos es de tamaño 2 porque cada registro de entrada en inputs tiene dos valores
bias = .123

# Para cada entrada ejecuta el procedimiento del perceptrón y guarda la salida
for i in range(targets.shape[0]):
    h = np.dot(inputs[i], weights) + bias
    y = escalon_unitario(h)
    outputs.append([h, y])

# Imprime la salida
print('Input 1', '  Input 2', '  Combinación lineal', '  Target', '       Output')
for i in range(4):
    print(inputs[i][0],'           ', inputs[i][1],'         ', (round(outputs[i][0],2)),'          ', targets[i], '          ', outputs[i][1])
```

**Compuerta XOR**


---



Para poder desarrollar una compuerta XOR es necesario implementar una red multicapa debido a que un solo perceptron no es capaz de dividir dos secciones.

```python
import numpy as np
import matplotlib.pyplot as plt

def escalon_unitario(x):
    return 1 if x >= 0 else 0

# Datos
inputs = np.array([[0,0],[0,1],[1,0],[1,1]])
targets = np.array([0,1,1,1])  # OR

# Inicialización
weights = np.random.rand(2)
bias = np.random.rand(1)

learning_rate = 0.1
epochs = 10

print("\nPesos iniciales:", weights)
print("Bias inicial:", bias)

errores_por_epoca = []

# Entrenamiento
for epoch in range(epochs):
    error_total = 0
    salidas_epoca = []

    print(f"\n===== ÉPOCA {epoch+1} =====")
    print("x1  x2   target   salida   error")

    for i in range(len(inputs)):
        x = inputs[i]
        target = targets[i]

        # Cálculo de salida
        h = np.dot(x, weights) + bias
        y = escalon_unitario(h)

        # Error
        error = target - y
        error_total += abs(error)

        # Guardar salida
        salidas_epoca.append(y)

        # Actualización
        weights = weights + learning_rate * error * x
        bias = bias + learning_rate * error

        print(f"{x[0]}   {x[1]}      {target}        {y}        {error}")

    print(f"Error total en época {epoch+1}: {error_total}")
    print(f"\nPesos epoca:  {epoch+1} :", weights)
    print(f"Bias epoca  {epoch+1}:", bias )
    
    errores_por_epoca.append(error_total)

print("\n===== RESULTADO FINAL =====")
for i in range(len(inputs)):
    h = np.dot(inputs[i], weights) + bias
    y = escalon_unitario(h)
    print(f"Entrada: {inputs[i]} -> Salida: {y}")

plt.plot(range(1, epochs+1), errores_por_epoca, marker='o', linestyle='-')
plt.title('Cambio del Error Total en cada epocas')
plt.xlabel('Epoca')
plt.ylabel('Error Total')
plt.show()
```
