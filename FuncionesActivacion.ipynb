```python
import numpy as np
```

```python
def escalon_unitario(x):
    if x >= 0:
        return 1
    else:
        return 0
```

```python
def sigmoide(x):
    sig = 1/(1+np.exp(-x))
    return sig
```

```python
def tangenteh(x):
    tan = np.tanh(x)
    return tan
```

```python
def relu(x):
    return max(0,x)
```
