```python
import numpy as np
```

```python
def escalon_unitario(x):
    if x >= 0:
        return 1
    else:
        return 0
```

```python
def sigmoide(x):
    sig = 1/(1+np.exp(-x))
    return sig
```

```python
def tangenteh(x):
    tan = np.tanh(x)
    return tan
```

```python
def relu(x):
    return max(0,x)
```

```python
#OR
inputs  = np.array([[0,0],[0,1],[1,0],[1,1]])
targets = np.array([0,     1,    1,    1])
outputs = []
weights = np.array([1,-1])
bias = -0.5

for i in range (targets.shape[0]):
    h = np.dot(inputs[i], weights) + bias
    y = escalon_unitario(h)
    outputs.append([h,y])

print('Input 1', 'Input 2', 'Conbinacion Lineal', 'Target', 'Output')
for i in range (4):
    print(inputs[i][0],'     ',inputs[i][1],'     ', (round(outputs[i][0],2)), '              ', targets[i], '     ', outputs[i][1])    
    
#Input 1 Input 2 Conbinacion Lineal Target Output
#0       0       -0.5                0       0
#0       1       -1.5                1       0
#1       0       0.5                 1       1
#1       1       -0.5                1       0   
```
